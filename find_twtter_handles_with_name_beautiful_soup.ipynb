```python
import pandas as pd
```

```python
df = pd.read_excel(r"/content/drive/MyDrive/candidates_ysk_sample.xlsx")
```

```python
df.reset_index(drop=True)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
df['TWITTER_HANDLE'] = ''
base_url = "https://www.google.com/search?q="
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}
for index, row in df.iterrows():
    name = f"{row['name']} {row['surname']}"
    city = row['district_slug']
    query = f"{name} milletvekili aday {city} Twitter"
    response = requests.get(base_url + query, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    for a in soup.find_all('a', href=True):
        if "twitter.com/" in a['href']:
            handle = a['href'].split("/")[3].split("?")[0]
            df.at[index, 'TWITTER_HANDLE'] = handle
            break
```

```python
df['SEARCH_URL'] = df['TWITTER_HANDLE'].apply(lambda x: f"https://www.twitter.com/{x}" if x else "")
```
